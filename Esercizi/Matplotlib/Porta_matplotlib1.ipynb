a. crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di
ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python
import pandas as pd 
date = {"last_name": ["Boyer","Copeland","Daniel","Serrano","Terrell","Vargas"],"orders_number":[185,50,196,58,24,28]}  #dictionari di liste
df = pd.DataFrame(date)
df
```

b. visualizzi un grafico a barre verticali per rappresentare i dat. Il risultato deve essere questo:

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes()
ax.bar(df["last_name"],df["orders_number"])        #creare grafico a barre
```

c. visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df["last_name"],df["orders_number"],color="r")
fig.suptitle('Ordini per ogni commesso')    #aggiungere un titolo 
```

d. Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
fig = plt.figure()
ax = plt.axes()
ax.bar(df.sort_values("orders_number", ascending = False)["last_name"],df["orders_number"].sort_values(ascending = False),color="g")
```

e. visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:

```python
fig = plt.figure()
ax = plt.axes()
ax.barh(df["last_name"],df["orders_number"])      #barh per girare il grafico a barre in orizziontale
```

f. crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente
dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in
ordine alfabetico sul cognome del componente dello staff. Il risultato deve essere questo:

```python
df2 = pd.DataFrame()
df2["last_name"] = df["last_name"].sort_values(ascending = True)
df2["orders_perc"] = df["orders_number"]/df["orders_number"].sum()*100
df2 
```

g. visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:

```python
plt.rcParams.update({'font.size': 14})    

fig = plt.figure(figsize=[10,10])
ax = plt.axes()

ax.pie(df2["orders_perc"], labels=df2["last_name"], startangle=90 , autopct='%1.1f%%')  #creare un grafico a torta

plt.show()
```

h. visualizzi lo stesso grafico ma con altri colori delle percentuali e un titolo appropriato al grafico

```python
plt.rcParams.update({'font.size': 14})     

fig = plt.figure(figsize=[10,10])   
ax = plt.axes()

ax.pie(df2["orders_perc"], labels=df2["last_name"], colors=["m","r","c","g","c","g"], startangle=90 , autopct='%1.1f%%')
fig.suptitle('Percentuale Ordini per ogni commesso', color='k')    #aggiungere il titolo 
plt.show()
```

i. rappresentare le stesse informazioni utilizzando come tipo di grafico il TreeMap (google it…)

```python
import squarify
squarify.plot(sizes=df2["orders_perc"], label=df2["last_name"], color=["m","r","r","b","c","g"], alpha=0.5 ) #crea un grafico TreeMap
plt.axis('off')   #toglie gli assi
plt.show()
```
