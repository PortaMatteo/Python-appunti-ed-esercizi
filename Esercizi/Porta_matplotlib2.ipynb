a. creare un dataframe contenente le informazioni riportate qui sotto, relative a quanti prodotti ha
venduto (‘quantity’) e quanto ha venduto in termini di $ (‘total_amount’) ogni membro dello staff
(colonna “last_name”):

```python
import pandas as pd 
date = {"last_name":["Boyer","Copeland","Daniel","Serrano","Terrell","Vargas"], "quantity": [766,213,812,269,100,114],"total_amount": [643049.37,203549.39,744259.20,253183.68,93851.95,123156.90]}
df = pd.DataFrame(date)
df
```

b. visualizzi due grafici a barre affiancati che indichino quantità vendute e valore venduto per ogni
dipendente. Il risultato deve essere questo:

```python
import matplotlib.pyplot as plt
fig, (ax1,ax2)= plt.subplots(1,2)       #creare due grafici uno di fianco all'altro
fig.set_size_inches(15,8)              #impostare grandezza grafici 
plt.rcParams.update({'font.size': 10})   #impostare grandezza caratteri
ax1.bar(df["last_name"],df["quantity"])
ax2.bar(df["last_name"],df["total_amount"])
```

c. visualizzi i grafici precedenti con un titolo, gli assi e la griglia.

```python
fig, (ax1,ax2)= plt.subplots(1,2)
fig.set_size_inches(25,12)
plt.rcParams.update({'font.size': 20}) 
ax1.bar(df["last_name"],df["quantity"])
ax2.bar(df["last_name"],df["total_amount"])
fig.suptitle("instogrammi")    #aggiungere un titolo 
ax1.set_title("Prodotti venduti")   #aggiungere titolo per ogni grafico
ax2.set_title("Guadagno")
ax1.set_xlabel("last_name")    #aggiungere asse
ax1.set_ylabel("quantity")
ax2.set_xlabel("last_name")
ax2.set_ylabel("total_amount")
ax1.grid()    #aggiungere la griglia
ax2.grid()
```

d. visualizzi, senza creare nuovi dataframe, il costo medio dei prodotti venduti da ogni dipendente. Il
risultato deve essere questo (prestare attenzione ai valori dell’asse delle ordinate):

```python
fig = plt.figure()
fig.set_size_inches(10,8)
ax = plt.axes()
ax.bar(df["last_name"],df["total_amount"]/df["quantity"])
ax.set_ylim([800, 1100])  #cambiare asse delle y 
```
