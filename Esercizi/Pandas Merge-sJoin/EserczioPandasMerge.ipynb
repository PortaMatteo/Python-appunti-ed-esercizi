- Visualizzare il nome della provincia e il nome del comune 

```python
comuni = geopandas.read_file("/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Com01012021_g")
comuni
```

```python
df_com_prov = pd.merge(province,comuni, how="left", on=["COD_PROV"])
df_com_prov = df_com_prov[["DEN_UTS","COMUNE"]]
df_com_prov
```

- Visualizzare il nome del comune e il nome della regione

```python
df_com_reg = pd.merge(comuni,regioni, how="left", on=["COD_REG"])
df_com_reg = df_com_reg[["COMUNE","DEN_REG"]]
df_com_reg
```

- Visualizzare numero di comuni per ogni provincia
